```python
#"In the name of you...."
import numpy as np
from scipy import io
import matplotlib.pyplot as plt
from skimage.transform import resize
```

```python
def dataset_hoda(n_X_train=1000,n_X_test=200,size_image=5):
    dataset_Hoda=io.loadmat("./dataset/Data_hoda_full.mat")

    X_train=np.squeeze(dataset_Hoda["Data"][:n_X_train])
    label_X_train=np.squeeze(dataset_Hoda["labels"][:n_X_train])
    X_test=np.squeeze(dataset_Hoda["Data"][n_X_train:n_X_train+n_X_test])
    label_X_test=np.squeeze(dataset_Hoda["labels"][n_X_train:n_X_train+n_X_test])

    X_train_resized=[resize(image,(size_image,size_image)) for image in X_train]
    X_train=[matrix.reshape(size_image**2) for matrix in X_train_resized]
    X_test_resized=[resize(image,(size_image,size_image))for image in X_test]
    X_test=[matrix.reshape(size_image**2) for matrix in X_test_resized]
    return X_train , label_X_train , X_test , label_X_test
# print(dataset_hoda())
```
